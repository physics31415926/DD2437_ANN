```python
from util import *
from rbm import RestrictedBoltzmannMachine 
from dbn import DeepBeliefNet
import matplotlib.pyplot as plt

image_size = [28,28]
train_imgs,train_lbls,test_imgs,test_lbls = read_mnist(dim=image_size, n_train=60000, n_test=10000)
''' restricted boltzmann machine '''

print ("\nStarting a Restricted Boltzmann Machine..")
Num_hidden_nodes = [200+100*i for i in range(4)]
mse=[]
for i in Num_hidden_nodes:
    rbm = RestrictedBoltzmannMachine(ndim_visible=image_size[0]*image_size[1],
                                        ndim_hidden=i,
                                        is_bottom=True,
                                        image_size=image_size,
                                        is_top=False,
                                        n_labels=10,
                                        batch_size=20,

    )
    mse.append(rbm.cd1(visible_trainset=train_imgs, n_iterations=20))
```

```python
for idx,i in enumerate(Num_hidden_nodes):
    plt.plot(mse[idx],label="%d hidden nodes"%i)
plt.title("mse over iterations with different hidden nodes")
plt.ylabel("MSE")
plt.xlabel("Epoch")
plt.legend()
plt.show()
```
