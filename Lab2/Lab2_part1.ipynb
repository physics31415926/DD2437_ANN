In general, it can get better performance on test set if we set more RBF nodes. But number of RBF nodes should be more than the dimension of input space (in this case, the dimension of input space is 1) and less than the number of training examples.
The width of RBFs can control the effect of each center on its neighbouring points. If sigma is small, the centers can affect the points in a bigger area. If sigma is large, the centers can affect the points just in a very small area. We should choose reasonable width of RBFs according to the number of RBF nodes, e.g.,when we have more nodes, we should decrease the width.
