```python
from Net import *
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

generate data

```python
x = np.linspace(-5,5,20)
y = np.linspace(-5,5,20)
X, Y = np.meshgrid(x, y)
Z = np.exp(-0.1*np.square(X)-0.1*np.square(Y))-0.5
ax = plt.axes(projection="3d")
ax.plot_wireframe(X, Y, Z, color='blue')
plt.show()
```

```python
targets = np.asarray(Z).reshape(400,1)
patternx = np.asarray(X).reshape(400,)
patterny = np.asarray(Y).reshape(400,)
patterns = np.ones((400,2))
patterns[:,0]=patternx
patterns[:,1]=patterny
```

```python
layers =[
    {'type':'linear','shape':(2,16)},
    {'type':'tanh'},
    {'type':'linear','shape':(16,16)},
    {'type':'tanh'},
    {'type':'linear','shape':(16,1)},
    {'type':'tanh'}
]
net = Net(layers)
opt = SGD(net.parameters,lr=1e-2)
net.train(patterns,targets,opt,epochs=10000)
```

```python
predicts = net.predict(patterns)
out = predicts.reshape(20,20)
ax = plt.axes(projection="3d")
ax.plot_wireframe(X, Y, out, color='blue')
plt.show()
```

## question 1


```python
sums1 = []
#shuffle
permute = list(range(400))
np.random.shuffle(permute)
patterns = patterns[permute, :]
targets = targets[permute]

trains = patterns[0:320,:]
targets_train = targets[0:320,:]
for n_nodes in range(1,26):
    layers =[
    {'type':'linear','shape':(2,n_nodes)},
    {'type':'tanh'},
    {'type':'linear','shape':(n_nodes,n_nodes)},
    {'type':'tanh'},
    {'type':'linear','shape':(n_nodes,1)},
    {'type':'tanh'}
    ]
    net = Net(layers)
    opt = SGD(net.parameters,lr=1e-2)
    net.train(trains,targets_train,opt,epochs=3000)
    predicts = net.predict(patterns)
    sum=0
    for ri in range(400):
        sum=sum+np.square(predicts[ri]-targets[ri])
    sums1.append(sum)
```

```python
n_nodes=[i for i in range(1,26)]
plt.plot(n_nodes,sums1)
plt.xlabel('number of nodes')
plt.ylabel('sum square error')
plt.title('effect of number of nodes')
plt.show()

print('best number of nodes:'+str(min(sums1)))
```
